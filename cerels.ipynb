```python
#overoll distribution of the data
plt.figure(figsize=(8,6))
plt.title("Overall Cereals Data Distribution")
sns.histplot(data = df)      #positively skewed
```

```python
#going to create distribution chart for necessary columns in data
# Lets create the function
def plot_histogram(column_data, column_name):
    plt.figure(figsize=(6, 3))
    plt.title(f"Distribution of {column_name}")
    sns.histplot(column_data, kde=True)
    plt.show()
```

```python
plot_histogram(df['type'], 'hot or cold type nutrients') #inference;majority likes to take cerels by cold but few likes hot
```

```python
plot_histogram(df['mfr'], 'manufactures')     #inference; K and G manufactered more
```

```python
plot_histogram(df['calories'],'calories' ) #inference;normally distributed  ,majority serves 100-120 gram per serving
```

```python
plot_histogram(df['protein'], 'protien')  #inference;majority serves 2-3 gram of protien per serving,some cerels have 6 gram per seving hence outlier
```

```python
plot_histogram(df['carbo'], 'carbo')   #inference;10-20 grams of carbo per serving
```

```python
plot_histogram(df['fat'], 'fat')   #inference;majority serves  0-1 gram per serving
```

```python
plot_histogram(df['sodium'], 'sodium')   #inference;normally disributed 150-250 gram per serving
```

```python
plot_histogram(df['sugars'], 'sugars') #inference;uniformly disributed
```

```python
plot_histogram(df['potass'], 'potass')    #inference;majority serves in between 0-150 milligrams per serving
```

```python
plot_histogram(df['vitamins'], 'vitamins')   #inference;majority serves 10,25,100 unit per serving
```

```python
plot_histogram(df['shelf'], 'shelf')#inference;majority has a shelf life between 2.75-3.00
```

```python
plot_histogram(df['weight'], 'weight')   #inference;major cerels weight in between 1.0-1.2 gram
```

```python
plot_histogram(df['cups'], 'cups') #inference;majority serves in between 0.8-1 cups per serving
```

```python
plot_histogram(df['rating'], 'rating')   #inference;major customers rated between 30-40,with 90 as outlier
```

```python
#visualisation
#q1:is there any relashionship between customer rating and calorie performance of cereals
x=df.calories
y=df.rating
plt.scatter(x, y, color='blue')
plt.xlabel("calories")
plt.ylabel("customer rating")
plt.title("relationship between customer rating and calorie performance")
plt.show()
```

```python
#which brand has the highest rating
x=df.name
y=df.rating
plt.figure(figsize=(5, 3))
plt.bar(x, y)
plt.xlabel("name of the brand")
plt.ylabel("customer rating")
plt.title("brand vs rating")
plt.show()
```
